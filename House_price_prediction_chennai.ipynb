```python
x = df_train.iloc[:,0:47]
x
```

```python
y = df_train.iloc[:,47]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state = 8)
```

```python
from sklearn.preprocessing import StandardScaler #Standardization
func = StandardScaler()
func.fit(x_train)
transf_train = func.transform(x_train)
transf_test = func.transform(x_test)
```

```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LinearRegression

spl = KFold(n_splits = 10)
model_1  = LinearRegression()
result = cross_val_score(model_1,x,y,cv = spl,scoring = 'neg_mean_absolute_error')
result

```

```python
model_1.fit(x_train, y_train)
```

```python
intercept = model_1.intercept_
coefficients = model_1.coef_

print('intercept:', model_1.intercept_)
print('coefficients:', model_1.coef_)
```

```python
print('r-squared: ', model_1.score(x_train, y_train))
print('r-squared: ', model_1.score(x_test, y_test))
```

```python
from sklearn.linear_model import SGDRegressor

model_2  = SGDRegressor(penalty='l1')
model_2.fit(x_train,y_train)
```

```python
intercept = model_2.intercept_
coefficients = model_2.coef_

print('intercept:', model_2.intercept_)
print('coefficients:', model_2.coef_)
```

```python
print('r-squared: ', model_2.score(x_train, y_train))
print('r-squared: ', model_2.score(x_test, y_test))
```

```python
model_2.n_iter_
```

```python
preds = model_2.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,preds)
```
